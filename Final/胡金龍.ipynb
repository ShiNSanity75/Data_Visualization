```python
hu15_2018_good = hu15_2018_NEW.loc[0:12]
hu15_2018_good
```

```python
fly1 = hu15_2018_good['飛球']/hu15_2018_good['出局數']
ground1 = hu15_2018_good['滾地球']/hu15_2018_good['出局數']
SO1 = hu15_2018_good['三振']/hu15_2018_good['出局數']

hu15_2018_good = hu15_2018_good.copy()
hu15_2018_good['飛球率'] = fly1
hu15_2018_good['滾地球率'] = ground1
hu15_2018_good['三振率'] = SO1
hu15_2018_good
```

```python
fig = plt.figure(figsize=(25,10))
ax1 = fig.add_subplot(111)

for (i, row) in hu15_2018_good.iterrows():
    plt.bar([i-0.2, i, i+0.2], [row['飛球率'], row['滾地球率'], row['三振率']],
             color=['#0066FF', '#77FF00', '#FF5511'], width=0.2, align='center', edgecolor='#000000')
    
plt.xticks(range(0, 13), hu15_2018_good['投手'], fontsize=23)

#plt.legend()
plt.title("胡金龍對投手出局方式比較(高打擊率)", fontsize=25)
plt.xlabel('NAME', fontsize=23)
plt.ylabel('比例', fontsize=23)
```

```python
hu15_2018_bad = hu15_2018_NEW.loc[22:31]
hu15_2018_bad = hu15_2018_bad.reset_index(drop=True)
hu15_2018_bad
```

```python
fly2 = hu15_2018_bad['飛球']/hu15_2018_bad['出局數']
ground2 = hu15_2018_bad['滾地球']/hu15_2018_bad['出局數']
SO2 = hu15_2018_bad['三振']/hu15_2018_bad['出局數']

hu15_2018_bad = hu15_2018_bad.copy()
hu15_2018_bad['飛球率'] = fly2
hu15_2018_bad['滾地球率'] = ground2
hu15_2018_bad['三振率'] = SO2
hu15_2018_bad
```

```python
fig = plt.figure(figsize=(25,10))
ax1 = fig.add_subplot(111)

for (i, row) in hu15_2018_bad.iterrows():
    plt.bar([i-0.2, i, i+0.2], [row['飛球率'], row['滾地球率'], row['三振率']],
             color=['#0066FF', '#77FF00', '#FF5511'], width=0.2, align='center', edgecolor='#000000')
    
plt.xticks(range(0, 10), hu15_2018_bad['投手'], fontsize=23)

#plt.legend()
plt.title("胡金龍對投手出局方式比較(低打擊率)", fontsize=25)
plt.xlabel('NAME', fontsize=23)
plt.ylabel('比例', fontsize=23)
```

```python
f1 = hu15_2018_good['飛球'].sum()/hu15_2018_good['出局數'].sum()
g1 = hu15_2018_good['滾地球'].sum()/hu15_2018_good['出局數'].sum()
S1 = hu15_2018_good['三振'].sum()/hu15_2018_good['出局數'].sum()

f2 = hu15_2018_bad['飛球'].sum()/hu15_2018_bad['出局數'].sum()
g2 = hu15_2018_bad['滾地球'].sum()/hu15_2018_bad['出局數'].sum()
S2 = hu15_2018_bad['三振'].sum()/hu15_2018_bad['出局數'].sum()
S1
```

```python
OUT_list = [{"類型":"高打擊率","飛球率":f1, "滾地球率":g1, "三振率":S1},
          {"類型":"低打擊率","飛球率":f2, "滾地球率":g2, "三振率":S2}]
OUT_df = pd.DataFrame(OUT_list)
OUT_df
```

```python
fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(111)

for (i, row) in OUT_df.iterrows():
    plt.bar([i-0.2, i, i+0.2], [row['飛球率'], row['滾地球率'], row['三振率']],
             color=['#0066FF', '#77FF00', '#FF5511'], width=0.2, align='center', edgecolor='#000000')
    
plt.xticks(range(0, 2), OUT_df['類型'], fontsize=23)

#plt.legend()
plt.title("胡金龍對投手出局方式比較(平均)", fontsize=25)
#plt.xlabel('NAME', fontsize=23)
plt.ylabel('比例', fontsize=23)
```
