```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

batter_2018 = pd.read_csv('batter_2018.csv',encoding='cp950')
#batter_2018 =open('batter_2018.csv')
#<_io.TextIOWrapper name='batter_2018.csv' mode='r' encoding='cp950'>編碼為encoding='cp950'

batter_Lamigo = pd.read_csv('batter_Lamigo.csv',encoding='cp950')
batter_Brothers = pd.read_csv('batter_Brothers.csv',encoding='cp950')
batter_Lions = pd.read_csv('batter_Lions.csv',encoding='cp950')

pitcher_2018 = pd.read_csv('pitcher_2018.csv',encoding='cp950')
pitcher_Lamigo = pd.read_csv('pitcher_Lamigo.csv',encoding='cp950')
pitcher_Brothers = pd.read_csv('pitcher_Brothers.csv',encoding='cp950')
pitcher_Lions = pd.read_csv('pitcher_Lions.csv',encoding='cp950')

batter_2018
#用pandas直接抓
#url = 'http://www.cpbl.com.tw/web/team_playergrade.php?&team=B04&gameno=01'
#table = pd.read_html(url)[0]
#table
```

```python
#這段在取出打數大於100的選手之打擊率擊上壘率(全年度)

#batter_2018.sort_values(by=['PA'],ascending=False)#以打數由大到小進行排序
#batter_2018[['NAME','PA']]
batter_2018_PA = batter_2018.loc[(batter_2018['PA'] > 100) , ['NAME','PA','AVG','OBP']]
batter_2018_PA = batter_2018_PA.reset_index(drop=True)
batter_2018_PA
```

```python
#分別取出打數大於100的選手之打擊率擊上壘率(對桃猿、兄弟、統一)
batter_Lamigo_PA = batter_Lamigo.loc[(batter_2018['PA'] > 100) , ['NAME','PA','AVG','OBP']]
batter_Lamigo_PA.rename(columns={"PA": "PA_Lamigo","AVG":"AVG_Lamigo","OBP":"OBP_Lamigo"}, inplace = True)
batter_Lamigo_PA = batter_Lamigo_PA.reset_index(drop=True)

batter_Brothers_PA = batter_Brothers.loc[(batter_2018['PA'] > 100) , ['NAME','PA','AVG','OBP']]
batter_Brothers_PA.rename(columns={"PA": "PA_Brothers","AVG":"AVG_Brothers","OBP":"OBP_Brothers"}, inplace = True)
batter_Brothers_PA = batter_Brothers_PA.reset_index(drop=True)

batter_Lions_PA = batter_Lions.loc[(batter_2018['PA'] > 100) , ['NAME','PA','AVG','OBP']]
batter_Lions_PA.rename(columns={"PA": "PA_Lions","AVG":"AVG_Lions","OBP":"OBP_Lions"}, inplace = True)
batter_Lions_PA = batter_Lions_PA.reset_index(drop=True)
batter_Lions_PA
```

```python
#合併四張表格並且以全年打數排序
batter_inner = pd.merge(batter_2018_PA, batter_Lamigo_PA, how='inner')
batter_inner = pd.merge(batter_inner, batter_Brothers_PA, how='inner')
batter_inner = pd.merge(batter_inner, batter_Lions_PA, how='inner')
batter_inner = batter_inner.sort_values(by=['PA'],ascending=False)
batter_inner = batter_inner.reset_index(drop=True)
batter_inner
```

```python
fig = plt.figure(figsize=(25,10))
ax1 = fig.add_subplot(111)

for (i, row) in batter_inner.iterrows():
    plt.bar([i - 0.3, i - 0.1, i + 0.1, i + 0.3], [row['AVG'], row['AVG_Lamigo'], row['AVG_Brothers'], row['AVG_Lions']],
            color=['#888888', '#008AB8', '#FFD700', '#FF8C00'], width=0.2, align='center', edgecolor='#000000')

plt.xticks(range(0, 17), batter_inner['NAME'], fontsize=23)

#for x,y in enumerate(batter_inner['AVG']):
   #plt.text(x, y, '%1.3f' % y, ha='center', va= 'bottom',fontsize=11)


plt.xlabel('NAME')
plt.ylabel('AVG')
```

```python
#取出所有投球局數大於15局之選手的自責分率及每局被上壘率(全年度)
pitcher_2018_IP = pitcher_2018.loc[(pitcher_2018['IP'] > 15) , ['NAME','IP','ERA','WHIP']]
#pitcher_2018_IP = pitcher_2018_IP.sort_values(by=['IP'],ascending=False)
pitcher_2018_IP = pitcher_2018_IP.reset_index(drop=True)
#pitcher_2018_IP.loc[(pitcher_2018_IP['NAME'] == '伍鐸')]
pitcher_2018_IP
```

```python
#分別取出投球局數大於15局之選手的自責分率及每局被上壘率(對桃猿、兄弟、統一)
pitcher_Lamigo_IP = pitcher_Lamigo.loc[(pitcher_2018['IP'] > 15) , ['NAME','IP','ERA','WHIP']]
pitcher_Lamigo_IP.rename(columns={"IP": "IP_Lamigo","ERA":"ERA_Lamigo","WHIP":"WHIP_Lamigo"}, inplace = True)
pitcher_Lamigo_IP = pitcher_Lamigo_IP.reset_index(drop=True)

pitcher_Brothers_IP = pitcher_Brothers.loc[(pitcher_2018['IP'] > 15) , ['NAME','IP','ERA','WHIP']]
pitcher_Brothers_IP.rename(columns={"IP": "IP_Brothers","ERA":"ERA_Brothers","WHIP":"WHIP_Brothers"}, inplace = True)
pitcher_Brothers_IP = pitcher_Brothers_IP.reset_index(drop=True)

pitcher_Lions_IP = pitcher_Lions.loc[(pitcher_2018['IP'] > 15) , ['NAME','IP','ERA','WHIP']]
pitcher_Lions_IP.rename(columns={"IP": "IP_Lions","ERA":"ERA_Lions","WHIP":"WHIP_Lions"}, inplace = True)
pitcher_Lions_IP = pitcher_Lions_IP.reset_index(drop=True)

pitcher_Lions_IP
```

```python
#合併四張表格並且以全年投球局數排序
pitcher_inner = pd.merge(pitcher_2018_IP, pitcher_Lamigo_IP, how='inner')
pitcher_inner = pd.merge(pitcher_inner, pitcher_Brothers_IP, how='inner')
pitcher_inner = pd.merge(pitcher_inner, pitcher_Lions_IP, how='inner')
pitcher_inner = pitcher_inner.sort_values(by=['IP'],ascending=False)
pitcher_inner = pitcher_inner.reset_index(drop=True)
pitcher_inner
```
